```python
import pandas as pd
import numpy as np
```

```python
#to read the data 
df=pd.read_csv("Data/gemstone.csv")
df.head()
```

```python
#we can remove the id column
df=df.drop(labels='id',axis=1)
```

```python
#to split the dependent and target features
X=df.iloc[:,:-1]
y=df.iloc[:,-1:]

```

```python
#Split the Dependent variables based on the categorical and numerical features

#numerical_cols=X.columns(X.dtypes!='object')
#categorical_cols=X.columns(X.dtypes=='object')
```

```python
#another way
numerical_cols=X.select_dtypes(exclude='object').columns
categorical_cols=X.select_dtypes(include='object').columns
```

```python
numerical_cols
```

```python
categorical_cols
```

```python
# Define the custom ranking for each ordinal variable- order in the ascending, fair has rank 1 and ideal has 5


cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
#Create a pipeline to perform the various steps in data processing and scaling
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# To create numerical pipeline

num_pipeline=Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
        ]
        )

# To create categorical pipeline        

cat_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinal Encoder', OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
        ('scaler',StandardScaler())
    ]
)

#To apply the pipeline in the columns using transformer function

preprocessor=ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_cols),
    ('Cat_pipeline',cat_pipeline,categorical_cols)]
)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
```

```python
X_train
```

```python
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
#Model Training we are using regression-linear, lasso,ridge
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
#to evaluate the metrics of the regression model we need to define a function

def evaluate_model(true,predicted):
    mse=mean_squared_error(true,predicted)
    mae=mean_absolute_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_square=r2_score(true,predicted)
    return mae,rmse,r2_square
```

```python
#create a model training for all the regression model by first creating a dictionary
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet()
}

#TRAVERSE THE EACH MODEL BY FIRST CONVERTING DICTIONARY TO LIST AND TRAIN THE MODEL WITH THE TRAIN DATA

#create a list to print the respective model output
trained_model_list=[]
model_list=[]
r2_list=[]


for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #Prediction for the test data
    y_pred=model.predict(X_test)
    mae,rmse,r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    print("Model Training Performance")
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2_SCORE",r2_square)

    r2_list.append(r2_square)

    print('='*35)
    print('\n')
```
